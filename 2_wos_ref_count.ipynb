```python
import os
import pandas as pd
import gzip
import json

import time
```

In this notebook, I combine all reference data table (20xx_uwm_reference.csv) together to get the unique reference list

```python
folder_path = "../wos_json"
file_list = os.listdir(folder_path)
file_list
```

```python
ref_path = "result"
ref_list = []
for x in os.listdir(ref_path):
    if x.endswith('reference.csv'):
        ref_list.append(x)
ref_list
```

```python
file_2018 = []
for x in file_list:
    year = int(x[3:7])
    if year>=2018:
        file_2018.append(x)
len(file_2018)
```

```python
with gzip.open(folder_path+'/'+file_2018[0], mode="rt") as f:
    data = [json.loads(line) for line in f]
```

```python
data[0]
```

```python
data1 = pd.read_csv(ref_path+'/'+ref_list[0])
data2 = data1.groupby('id').count()['citing_paper'].reset_index()
data2 = data2.rename(columns={'citing_paper': "2018"})
```

```python
x = 0
for i in range(len(ref_list)):
    data1 = pd.read_csv(ref_path+'/'+ref_list[i])
    data2 = data1.groupby('id').count()['citing_paper'].reset_index()
    data2 = data2.rename(columns={'citing_paper': str(i+2018)})
    if x ==0:
        data = data2
        x = 1
    else:
        data = data.merge(data2,on='id',how='outer')
```

```python
data = data.fillna(0)
data[['2018','2019','2020','2021','2022']] = data[['2018','2019','2020','2021','2022']].astype(int)
data
```

```python
data.to_csv('reference_count.csv',index=False)
```
