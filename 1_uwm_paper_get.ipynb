```python
# Sorry, this part of code was lost, so I just rewrite some of them. As the code of getting different metadata is the same, so i just rewrite those for identifiers and use "..." to represent similar code

for i in range(2018,2023):
    file_i = []
    for x in file_list:
        year = int(x[3:7])
        if year==i:
            file_i.append(x)

    citing_file = open('result/'+str(i)+'_uwm_paper.csv','w')
    citing_file.write('id,title,Doi,issn,source_title,document_types,pub_year,pub_month......')
    reference_file = open('result/'+str(i)+'_uwm_reference.csv','w')
    reference_file.write('citing_paper','id', 'year', 'author', 'title', 'work', 'doi', 'page', 'volume', 'occurrence_order', 'physical_sections')
    for file in file_i:
        file_title = file[:-8]
        file_path = folder_path+'/'+file
        with gzip.open(file_path, mode="rt") as f:
            data = [json.loads(line) for line in f]

        for paper in data:
            address_number_list = []
            uwm_auther_list = []
            id = ''
            doi = ''
            issn = ''
            
            #filter for uwmadison paper
            if paper['addresses']==None:
                continue
            for address in paper['addresses']:
                if address['organizations']==None:
                    continue
                for org in address['organizations']:
                    if org == 'University of Wisconsin Madison':
                        address_number_list.append(address['number'])
            if address_number_list==[]:
                continue
            else:
                id = paper['id']
                if paper['identifiers']!=None:
                    for identifier in paper['identifiers']:
                        if identifier['type']=='doi':
                            doi = identifier['value']
                        if identifier['type']=='issn':
                            issn = identifier['value']
                # ..... coding for getting other metadata

                citing_file.write(id+','+paper['title']+','+doi+','+issn......)

                # get reference data
                if paper['references']!=None:
                    for reference in paper['references']:
                        reference_file.write(id+','+reference['id']+','+reference['year']......)
                        
```
